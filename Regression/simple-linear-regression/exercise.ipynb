SAT sering digunakan sebagai indikator kemampuan akademik awal, dan proyek ini mencoba mengukur sejauh mana skor SAT dapat menjadi prediktor linier terhadap prestasi akhir mahasiswa dalam bentuk GPA.

* Menguji apakah ada hubungan linier antara SAT scores dan GPA.

## Format Dataset
* Fitur input (X): SAT_Score (skor ujian SAT siswa)

## Stack & Tools
* Bahasa: Python
* Library: pandas, numpy, matplotlib, seaborn, scikit-learn

### Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Import dataset

```python
dataframe = pd.read_csv('dataset/gpa.csv')
# x = np.array([dataframe['Age']]).reshape(-1, 1)
# x = np.array([dataframe['Age']])
# x = dataframe.iloc[:, 0].values.reshape(-1,1)
x = dataframe.iloc[:, :-1].values

# ubah meter ke centimeter
# dataframe['Height'] = (dataframe['Height'] * 100).astype(int)
# y = np.array([dataframe['Height']])
y = dataframe.iloc[:, -1].values
# y = dataframe['Height']
# print(y)
```

## Splitting dataset into training and testing

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
# print(y_train)
```

## Training data with simple linear regression model

```python
from sklearn.linear_model import LinearRegression
regresi = LinearRegression()
regresi.fit(x_train, y_train)
```

## Make predict result

```python
y_pred = regresi.predict(x_test)
# print(y_pred, y_test)
```

# Visualize the training result

```python
# merah = data yg di training
plt.scatter(x_train, y_train, color='red')
x_predict_training = regresi.predict(x_train)
plt.plot(x_train, x_predict_training, color='blue')
plt.title('SBMPTN VS IPK (Training set)')

plt.xlabel('SBMPTN')
plt.ylabel('IPK')

plt.show()

```
