# Simple Linear Regression

## Import Necessary Library

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Import dataset

```python
df = pd.read_csv('../dataset/Salary_Data.csv')
x = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

## Split the dataset into the training data and testing data 

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

## Training the simple linear regression model on the training data 

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

lr.fit(x_train, y_train)
```

## Predict the test result

```python
y_pred = lr.predict(x_test)
```

## Visualising the training set result

```python
# red = data yg di training
plt.scatter(x_train, y_train, color='red')
x_predict_training = lr.predict(x_train)
plt.plot(x_train, x_predict_training, color='blue')
plt.title('Gaji VS Pengalaman (training set)')

plt.xlabel('Pengalaman Kerja')
plt.ylabel('Gaji')

plt.show()
```

## Visualising the predicted set result

```python
# red = data yg di training
plt.scatter(x_test, y_test, color='red')
x_predict_test = lr.predict(x_train)
plt.plot(x_train, x_predict_test, color='blue')
plt.title('Gaji VS Pengalaman (test set)')

plt.xlabel('Pengalaman Kerja')
plt.ylabel('Gaji')

plt.show()
```
