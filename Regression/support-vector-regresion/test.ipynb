# Support Vector Regression (SVR)

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the datasets

```python

dataset = pd.read_csv('dataset/Position_Salaries.csv')
x = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
print(x)
print(y)
```

```python
y = y.reshape(len(y), 1)
# print(len(y))
# print(y)
```

## Feacture Scalling 

```python
from sklearn.preprocessing import StandardScaler
StandScaler_x = StandardScaler()
StandScaler_y = StandardScaler()
x = StandScaler_x.fit_transform(x)
y = StandScaler_y.fit_transform(y)
```

```python
print(x)
```

```python
print(y)
```

## Training the SVR model on the whole dataset

```python
from sklearn.svm import SVR
regresi = SVR(kernel = 'rbf')
regresi.fit(x, y)
```

## Predict a new results

```python
StandScaler_y.inverse_transform(regresi.predict(StandScaler_x.transform([[6.5]])).reshape(-1,1))
```

## Visualising the SVR results

```python
plt.scatter(StandScaler_x.inverse_transform(x), StandScaler_y.inverse_transform(y), color='red')
plt.plot(StandScaler_x.inverse_transform(x), StandScaler_y.inverse_transform(regresi.predict(x).reshape(-1,1)), color = 'blue')
plt.title('Truth or Bluff (Support Vector Regression(SVR))')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

## Visualising the SVR results (for higher resolution and smoother curve)

```python
x_grid = np.arange(min(StandScaler_x.inverse_transform(x)), max(StandScaler_x.inverse_transform(x)), 0.1)
# print(x_grid)
x_grid = x_grid.reshape((len(x_grid), 1))
# print(x_grid)

plt.scatter(StandScaler_x.inverse_transform(x), StandScaler_y.inverse_transform(y), color='red')
plt.plot(x_grid, StandScaler_y.inverse_transform(regresi.predict(StandScaler_x.transform(x_grid)).reshape(-1,1)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```
